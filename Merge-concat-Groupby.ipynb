```python
import pandas as pd
```

## Create a DataFrame

```python
df1 = pd.DataFrame({'City':['Chennai','Mumbai','HYD','Delhi','Banglore'],
                    'Temperature': [23 ,34 ,45 ,21 ,39],
                    'Humidity': [67 ,69 ,70 ,88 ,67]
                   })
df1
```

```python
df2 = pd.DataFrame({
                   'City':['HYD','Nellore','Pune','Vizag','Kochhi'],
                    "Wind" : ["32 Km/h" ,"30 Km/h" , "40 Km/h" ,"23 Km/h" , "33 Km/h"]
                  })
df2
```

## Inner Join

```python
# Merge to tables using inner 
df3 = pd.merge(df1,df2, on='City') # By default is always Inner Join
df3
```

```python
df4 = pd.merge(df1,df2, on='City',how='inner') # second method
df4
```

## Outer Join

```python
df5 = pd.merge(df1,df2, on = 'City', how = 'outer') # with on condition
df5
```

```python
df5 = pd.merge(df1,df2, how = 'outer') # without on condition
df5
```

## Left Join

```python
df6 = pd.merge(df1,df2,on = 'City', how = 'left')
df6
```

```python
# Right Join
df7 = pd.merge(df1,df2,on = 'City', how = 'right')
df7
```

## Concatination

```python
df = pd.concat([df1,df2])
df
```

```python
df8 = pd.concat([df1,df2], ignore_index = True)
df8
```

```python
df9 = pd.concat([df1,df2],axis=1) # Adding the data column wise 
df9
```

```python
df10 = pd.concat([df1,df2],axis=0) #  adding the data row wise 
df10
```

```python
df8 = pd.concat([df1,df2], ignore_index = True)
df8
```

```python
path = "Glassdoor Gender Pay Gap.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.groupby('JobTitle')['Bonus'].sum()
```

```python
df.groupby('JobTitle')[['Bonus','BasePay']].agg(['mean','max','min','sum'])
```

```python
df.groupby('JobTitle').agg({'Bonus':['mean','max','min','sum'],'BasePay':['mean','max','min','sum']})
```

```python
df.groupby('JobTitle')['Gender'].value_counts()
```

```python
df.groupby('JobTitle')['Gender'].value_counts().unstack()
```

```python
df.groupby('JobTitle')[['Dept','Gender']].value_counts().unstack()
```

```python
df.groupby('JobTitle')[['Bonus','BasePay']].median()
```

```python
df.groupby('JobTitle')[['Bonus','BasePay']].std()
```
