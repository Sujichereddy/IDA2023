```python
import pandas as pd
import numpy as np
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
# Read data
df_old = pd.read_csv("Crimes in Denmark till 2009.csv")
df_old.describe(include ='all')
```

```python
# Read Data
df_new_part1 = pd.read_excel("Crimes in Denmark till 2019 part1.xlsx")
df_new_part1.describe(include = 'all')
```

```python
# Read Data
df_new_part2 = pd.read_csv("Crimes in Denmark till 2019 part2.txt")
df_new_part2.describe(include = 'all')
```

## Concatination

```python
df_concat_2019 = pd.concat([df_new_part1,df_new_part2],ignore_index= True)
df_concat_2019.head()
```

## Removing Duplicates

```python
df_concat_2019[df_concat_2019.duplicated(['REGION','TYPE OF OFFENCE'])]
```

```python
new_df = df_concat_2019.drop_duplicates(['REGION','TYPE OF OFFENCE'])
new_df.describe(include='all')
```

```python
df_old.head(1)
```

```python
new_df.head(1)
```

## Merge two dataframes

```python
df_merge = df_old.merge(new_df,on = ['REGION','TYPE OF OFFENCE'])
len(df_merge.columns)
```

```python
len(df_merge)
```

```python
df_merge[df_merge.duplicated(['REGION','TYPE OF OFFENCE'])]
```

```python
len(df_merge)
```

## Convert columns into rows

```python
data = df_merge.melt(id_vars = ['REGION','TYPE OF OFFENCE'], var_name ='Time',value_name ='value')
data.head()
```

## Check how many crime is reported per region

```python
data.groupby(['REGION'],as_index = False)['value'].sum()
```

## Check how many crime we have in database based on the Region and Type of offence

```python
data.groupby(['REGION','TYPE OF OFFENCE'],as_index = False)['value'].sum()
```

```python
data['year']= data.Time.apply(lambda x : int(x.split('Q')[0]))
data['Quarter']= data.Time.apply(lambda x : int(x.split('Q')[1]))
```

## Show a horizontal barchart for number of the crimes (TYPE OF OFFENCE)

```python
agg_crime = data.groupby(['TYPE OF OFFENCE']).agg({'value':'sum'})
agg_crime.plot(kind ='barh',figsize =(10,20))
```

## Create a linechart to show number of the 'Attempted homicide' and 'Homicide' from 2009 till 2019 (by two different lines)

```python
aggdata = data.groupby(['TYPE OF OFFENCE','year']).agg({'value':'sum'}).reset_index()
sb.lineplot(data = aggdata[aggdata['TYPE OF OFFENCE'].isin(['Attempted homicide','Homicide'])], x = 'year',y ='value',hue ='TYPE OF OFFENCE')
```

```python
sb.barplot(data = aggdata[aggdata['TYPE OF OFFENCE'].isin(['Attempted homicide','Homicide'])], x = 'year',y ='value',hue ='TYPE OF OFFENCE')
```
