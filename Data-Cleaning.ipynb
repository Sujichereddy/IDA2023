### Converting the string to upper case using LAMBDA

```python
person_new['Gender']=person_new['Gender'].fillna('').apply(lambda x: x.strip().upper())
person_new['Gender']
```

```python
person_new.Last_name = person_new.Last_name.apply(lambda x: x.strip().upper())
person_new.Last_name 
```

```python
person_new.Nationality = person_new.Nationality.apply(lambda x: x.strip().upper())
person_new.Nationality
```

### Coverting Datatype from Float to Integer

```python
person_new.Age = person_new.Age.fillna(0).astype(int)
person_new.Age
```

```python
person_new.Postal_code = person_new.Postal_code.fillna(0)
person_new.Postal_code = person_new.Postal_code.astype(int)

person_new
```
